```python
def sql_query(query):
    result = engine.execute(text(query))
    df = []
    [df.append(row) for row in result]
    df = pd.DataFrame(df)
    return df
```

<br><h1>1. Create a query or queries to extract the information you think may be relevant for building the prediction model. 
It should include some film features and some rental features. Use the data from 2005.<h1>

The main available features that can help us with the prediction in our case are the film category (or genre), the length and the rating. <br>Starring actors is not feasible as there would be multiple entries for each film, and special features and description are too complicated to encode for the linear model (in particular despcription is unique for each film which means there would be as many new columns for that value as films are, and including special features makes the model perform even worse, see at the end). <br>Therefore we will restrict our selection to the fields stated before.

```python
query1 = '''SELECT f.film_id, f.title, c.name as category, f.length, f.rating from film as f
JOIN inventory as i
ON f.film_id = i.film_id
JOIN film_category as fc
ON f.film_id = fc.film_id
JOIN category as c
on c.category_id = fc.category_id
JOIN rental as r
ON i.inventory_id = r.inventory_id
WHERE r.rental_date LIKE '2005%'
GROUP BY f.film_id, category
ORDER BY film_id;'''
```

```python
sql_query(query1)
```

<br><h1>2. Create a query to get the total amount of rentals in June for each film.<h1>

```python
query_june= '''SELECT f.film_id, f.title, c.name as category, f.length, f.rating, count(r.inventory_id) as number_of_rentals_june from film as f
JOIN inventory as i
ON f.film_id = i.film_id
JOIN film_category as fc
ON f.film_id = fc.film_id
JOIN category as c
on c.category_id = fc.category_id
JOIN rental as r
ON i.inventory_id = r.inventory_id
WHERE r.rental_date LIKE '2005-06%'
GROUP BY f.film_id, category
ORDER BY film_id;'''
```

```python
rentals_june = sql_query(query_june)
rentals_june
```

<br><h1>3. Do the same with July.<h1>

```python
query_july = '''SELECT f.film_id, f.title, c.name as category, f.length, f.rating, count(r.inventory_id) as number_of_rentals_july from film as f
JOIN inventory as i
ON f.film_id = i.film_id
JOIN film_category as fc
ON f.film_id = fc.film_id
JOIN category as c
on c.category_id = fc.category_id
JOIN rental as r
ON i.inventory_id = r.inventory_id
WHERE r.rental_date LIKE '2005-07%'
GROUP BY f.film_id, category
ORDER BY film_id;'''
```

```python
rentals_july = sql_query(query_july)
rentals_july
```

<br><h1>4. Create a new column containing (Yes/No) for each film whether or not the number of monthly rentals in July was bigger than in June. Your objective will be to predict this new column.<br><br>5. Read the data into a Pandas dataframe.<h1>

```python
rentals = pd.concat([rentals_june, rentals_july], axis = 0)
rentals
```

```python
rentals['number_of_rentals_june'] = rentals['number_of_rentals_june'].fillna(value=0)
rentals['number_of_rentals_july'] = rentals['number_of_rentals_july'].fillna(value=0)
rentals['comparison'] = np.where((rentals['number_of_rentals_july'] > rentals['number_of_rentals_june']), 'Yes', 'No')
rentals
```

```python
rentals['comparison'].value_counts()
```

```python
rentals.head()
```

<br><h1>6. Analyze extracted features and transform them. You may need to encode some categorical variables or scale numerical variables.<h1><br><h1>7. Create a logistic regression model to predict this new column from the cleaned data.<h1><br><h1>8. Evaluate the results.<h1>

```python
X = rentals.drop('comparison',axis = 1)
X = X.drop('number_of_rentals_june', axis = 1)
X = X.drop('number_of_rentals_july', axis = 1)
X = X.drop('title', axis=1)
y = rentals['comparison']
```

```python
X_cat = X.select_dtypes(object)
X_cat = pd.get_dummies(X_cat)
X_cat
```

```python
y = y.replace(to_replace=('Yes', 'No'), value= (1, 0))
```

```python
X_num = X.select_dtypes(np.number)
```

```python
transformer = StandardScaler()
transformer.fit(X_num)

X_num_nrm = transformer.transform(X_num)
X_num_nrm = pd.DataFrame(X_num_nrm, columns=X_num.columns, index=X_num.index)
X_num_nrm.head()
```

```python
X = pd.concat([X_cat, X_num_nrm], axis=1)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
lr = LogisticRegression(random_state=0, solver='lbfgs')
lr.fit(X_train, y_train)
```

```python
y_pred_test = lr.predict(X_test)

print("The accuracy in the TEST set is: {:.2f}".format(accuracy_score(y_test,y_pred_test)))
print("The precision in the TEST set is: {:.2f}".format(precision_score(y_test,y_pred_test)))
print("The recall in the TEST set is: {:.2f}".format(recall_score(y_test,y_pred_test)))
print("The F1 in the TEST set is: {:.2f}".format(f1_score(y_test,y_pred_test)))
print("The Kappa in the TEST set is: {:.2f}".format(cohen_kappa_score(y_test,y_pred_test)))
```

```python
cm_test = confusion_matrix(y_test,y_pred_test)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=lr.classes_)
disp.plot()
plt.show()
```

<br><h4>**Not a very good prediction is possible with this data. After all we are only considering the rentals from 2 different months.**<h4>

We will try again reducing the test size to 20% and increasing iterations to 1000 in the LR model, also changing the solver to liblinear (which according to docs works better for small linear models).

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
lr = LogisticRegression(random_state=0, max_iter=1000, solver='liblinear')
lr.fit(X_train, y_train)
```

```python
y_pred_test = lr.predict(X_test)

print("The accuracy in the TEST set is: {:.2f}".format(accuracy_score(y_test,y_pred_test)))
print("The precision in the TEST set is: {:.2f}".format(precision_score(y_test,y_pred_test)))
print("The recall in the TEST set is: {:.2f}".format(recall_score(y_test,y_pred_test)))
print("The F1 in the TEST set is: {:.2f}".format(f1_score(y_test,y_pred_test)))
print("The Kappa in the TEST set is: {:.2f}".format(cohen_kappa_score(y_test,y_pred_test)))
```

```python
cm_test = confusion_matrix(y_test,y_pred_test)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=lr.classes_)
disp.plot()
plt.show()
```

Not a big change (performance is actually worse), we can see that the problem is that the initial data is very reduced and not good for our model.
